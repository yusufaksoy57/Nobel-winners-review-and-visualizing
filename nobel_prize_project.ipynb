## The most Nobel of Prizes
<p>The Nobel Prize is perhaps the world's most well known scientific award. Except for the honor, prestige and substantial prize money the recipient also gets a gold medal showing Alfred Nobel (1833 - 1896) who established the prize. Every year it's given to scientists and scholars in the categories chemistry, literature, physics, physiology or medicine, economics, and peace. The first Nobel Prize was handed out in 1901, and at that time the Prize was very Eurocentric and male-focused, but nowadays it's not biased in any way whatsoever. Surely. Right?</p>
<p>Well, we're going to find out! The Nobel Foundation has made a dataset available of all prize winners from the start of the prize, in 1901, to 2016. Let's load it in and take a look.</p>

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np
# Reading in the Nobel Prize data
nobel = pd.read_csv("datasets/nobel.csv")

# Taking a look at the first several winners
nobel.head(5)
```

```python
#the number of (possibly shared) Nobel Prizes handed out between 1901 and 2016.
display(len(nobel))
#the number of prizes won by male and female recipients.
display(nobel["sex"].value_counts())
#the number of prizes won by the top 5 nationalities.
nobel["birth_country"].value_counts().head(5)
```

## USA dominance
<p>The most common Nobel laureate between 1901 and 2016 was a man born in the United States of America.</p>

```python
nobel.columns
```

```python
# The proportion of USA born winners per decade
nobel['usa_born_winner'] = (nobel['birth_country']=='United States of America')*100
nobel['decade'] = (np.floor(nobel['year']/10)*10).astype(int)
prop_usa_winners = nobel.groupby("decade",as_index=False)["usa_born_winner"].mean()

# The proportions of USA born winners per decade
display(prop_usa_winners)
```

##  Visualization of US dominance.

```python
sns.set()
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [11, 7]

# Plotting USA born winners 
ax = sns.lineplot(x="decade",y="usa_born_winner",data=nobel)

from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter(1))
```

## Above table is the gender of a typical Nobel Prize winner.

```python
nobel['female_winner'] = nobel["sex"] == "Female" 
prop_female_winners = nobel.groupby(["decade","category"],as_index = False)["female_winner"].mean()
ax = sns.lineplot(data=prop_female_winners,x="decade",y="female_winner",hue="category")
```

## The first Female to win the Nobel Prize

```python
a= nobel[nobel["sex"] == "Female"]
a.nsmallest(1,"year")
```

## Repeat laureates
<p> we can see which scientist won 2 or more awards </p>

```python
# Selecting the laureates that have received 2 or more prizes.
nobel.groupby(by= "full_name").filter(lambda x: len(x) >=2)
```

## How old are they when they get the prize?
<p>I want to see how old are you generally when you get the prize?</p>

```python
nobel.head(1)
```

```python
# Converting birth_date from String to datetime
nobel["birth_date"]=pd.to_datetime(nobel["birth_date"]) # Calculating the age of Nobel Prize winners
# If we subtract their age from the age at which they won the prize, we can find out how old they are.
nobel['age'] =  nobel["year"] - nobel["birth_date"].dt.year
# Plotting the age of Nobel Prize winners
sns.lmplot(data=nobel,x="year",y="age")
```

## Age differences between prize categories
<p>Let's look at age trends within different prize categories.</p>

```python
# Same plot as above, but separate plots for each type of Nobel Prize
sns.lmplot(data=nobel,x="year",y="age",row="category")
```

## Oldest and youngest winners

```python
# The oldest winner of a Nobel Prize as of 2016
display(nobel.nlargest(1,"age"))
# The youngest winner of a Nobel Prize as of 2016
nobel.nsmallest(1,"age")
```
